```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from motors.plot import prepare_plot
from motors.plot import paper_plot
prepare_plot()
```

```python
data = np.genfromtxt('scan-output.txt')
```

```python
fig = plt.figure(figsize=(6 * 1.2, 6))
ax = plt.subplot()
ax.plot(data[:,0], data[:,1])
ax.scatter(data[:,0], data[:,1], s=120)
ax.set_xlabel('Dihedral angle (degrees)')
ax.set_ylabel('Energy (Hartrees)')
paper_plot(fig)
```

```python
fig = plt.figure(figsize=(6 * 1.2, 6))
ax = plt.subplot()
ax.plot(data[:,0], data[:,1] * 630 - data[0, 1] * 630)
ax.scatter(data[:,0], data[:,1] * 630 - data[0, 1] * 630, s=120)
ax.set_xlabel('Dihedral angle (degrees)')
ax.set_ylabel('Energy (kcal/mol)')
paper_plot(fig)
```
